# 2. Visualisasi plot distribusi setiap kolom numerik.

```python
def visualize_col(col):
    f, [ax1, ax2] = plt.subplots(1,2,figsize=(13,5))
    f.tight_layout()
    f.set_facecolor('white')

    ax1.hist(df[col])
    ax1.set(title='Histogram Data ' + col, xlabel = col, ylabel='frekuensi')
    ax1.grid()

    ax2.boxplot(df['chlorides'], showmeans=True)
    ax2.set(title='Boxplot Data ' + col, ylabel='frekuensi')
    ax2.grid()
    
    f.subplots_adjust(wspace=0.2)

    plt.show()
```

```python
visualize_col('fixed acidity')
```

# 3. Normality Test Setiap Kolom Numerik.

```python
def normality_test(col):
    plt.figure(facecolor='white', figsize=(7,5))
    plt.hist(df[col], density=True)
    plt.title('Histogram Data ' + col)
    plt.grid()


    xmin, xmax = plt.xlim()
    ymin, ymax = plt.ylim()
    ymed = (ymin + ymax) / 2
    mu, std = stats.norm.fit(df[col]) 
    x = np.linspace(xmin, xmax, 100)
    p = stats.norm.pdf(x, mu, std)

    plt.plot(x, p, 'k', linewidth=2)


    z, pval = stats.normaltest(df[col])
    mmmm, nnnn = stats.normaltest(df['volatile acidity'])
    txt = 'p value = ' + str(round(pval,15))
    txt += '\np value = ' + str(round(nnnn,15))
    if pval < 0.05:
        txt += '\n\nData ' + col + '\ntidak berdistribusi normal'
    else:
        txt += '\n\nData ' + col + '\nberdistribusi normal'

    plt.text(1.1*xmax, ymed, txt, fontsize='large', color='red', family='monospace')

    plt.show()
```
